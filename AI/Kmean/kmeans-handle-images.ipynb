## Library import
- numpy
- matplotlib
- sklearn

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

## Show original image 

```python
img = plt.imread("tree.jpg")
plt.imshow(img)
plt.show()
```

The shape of the image -> tuple

```python
img.shape
```

## Image parameters
img.shape[0] = 656

img.shape[1] = 561

3 = 3D space vector

*img.reshape*: transform img -> list array [[1,2,3],[1,2,3]] -> used Kmeans algorithm

*reshape()*: function shapes an array without changing the data of the array

```python
height = img.shape[0]
width  = img.shape[1]
img    = img.reshape(height*width, 3)
img
```

## Train data

*n_clusters* <=> number colors

*fit*: train
    
*kmeans.predict(img)*: Labels of each point

*kmeans.cluster_centers_*: Coordinates of cluster centers

```python
#Clusters = 4 <=> 4 colors
kmeans = KMeans(n_clusters=4).fit(img)
labels = kmeans.predict(img)
clusters = kmeans.cluster_centers_
clusters
```

## Create data zero
Create an array of given shape and type as given array, with zeros of the same dimension as img

```python
img_zero = np.zeros_like(img)
img_zero
```

Set value for **img_zero** by clusters with position of labels

```python
for i in range(len(img_zero)):
    img_zero[i] = clusters[labels[i]]
img_zero
```

## Transform image

```python
img_zero = img_zero.reshape(height, width, 3)
img_zero
```

## Show image again

```python
plt.imshow(img_zero)
plt.show()
```
